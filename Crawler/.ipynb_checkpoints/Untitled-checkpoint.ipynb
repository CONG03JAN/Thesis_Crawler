```python
import pymongo    
import re

client = pymongo.MongoClient("localhost", 27017)
db = client.Thesis
dbCateList = db.CateList
```

```python
suggestions = []

results = dbCateList.find()

collection = results

instring = "红烧肉"

regexString = instring.split(" ")

pattern = '.*?'.join(regexString)    # Converts 'djm' to 'd.*?j.*?m'
print(pattern)
regex = re.compile(pattern)         # Compiles a regex.
# collection = ["汉语是世界上最优美的语言", "正则表达式是一个很有用的工具"]
for item in collection:
    match = regex.search(item['cateName'])      # Checks if the current item matches the regex.
    if match:
        suggestions.append(item['cateName'])

for item in suggestions:
    print(item)
```
